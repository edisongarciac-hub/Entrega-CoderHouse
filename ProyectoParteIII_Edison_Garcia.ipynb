# **2. Predicción del precio del los automoviles despues de 5 años, tomando en cuenta el valor marcado en el odometro y su precio actual.**

Este notebook implementa un flujo de trabajo de Machine Learning completo utilizando los datos de precio (sellingprice) y odómetro (odometer) para la predicción del precio futuro y una tarea de clasificación de precios.

**Importar Librerías**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Para Regresión
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# Para Clasificación
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

**Carga y Exploración de Datos**

```python
import pandas as pd

# 1. Coloca aquí el ID de tu archivo de Google Drive
file_id = '1f4V0KxbqJyzG9EfTwr_HTcJO7giTfHTk'

# 2. Construir la URL de descarga directa
# Esta es la estructura mágica que convierte un enlace de Drive en descarga
download_url = f'https://drive.google.com/uc?id={file_id}&export=download'

try:
    # Leer el archivo CSV con la URL corregida
    df = pd.read_csv(download_url, encoding='latin1')

    print("✅ Archivo cargado correctamente.")
    print("Primeras 5 filas:")
    print(df.head())

except Exception as e:
    # Este bloque solo se ejecutará si hay un problema después de la corrección
    print(f"❌ Ocurrió un error al leer el archivo. Verifica la configuración.")
    print(f"Error detallado: {e}")

print("## 2.1. Primeras filas")
print(df.head(3))

print("\n## 2.2. Información del DataFrame")
df.info()

print("\n## 2.3. Estadísticas descriptivas de las variables clave")
print(df[['odometer', 'sellingprice']].describe())

# Visualización de la relación entre variables clave
plt.figure(figsize=(8, 6))
sns.scatterplot(x='odometer', y='sellingprice', data=df)
plt.title('Relación entre Odométer y Precio de Venta Actual')
plt.show()
#
```

**Preparación de Datos**

Limpieza de Datos

```python
# Eliminar filas con valores nulos en las columnas clave
df_clean = df.dropna(subset=['odometer', 'sellingprice'])

# Asegurar que son tipos numéricos y eliminar valores no válidos (precios < 100)
df_clean['odometer'] = pd.to_numeric(df_clean['odometer'], errors='coerce')
df_clean['sellingprice'] = pd.to_numeric(df_clean['sellingprice'], errors='coerce')

df_clean = df_clean[
    (df_clean['sellingprice'] > 100) & (df_clean['odometer'] >= 0)
]

print(f"Filas después de la limpieza: {len(df_clean)}")
```

Calculo de depreciación a 5 años

Se define el valor a 5 años (Regresión): Se usa el factor de 0.50 (50% de valor restante) como una estimación promedio de depreciación en 5 años.

```python
# Definición
DEPRECIATION_FACTOR = 0.50
df_clean['future_price_5yr'] = df_clean['sellingprice'] * DEPRECIATION_FACTOR

# Definir variables (X = odómetro, y = precio futuro)
X = df_clean[['odometer']].values.reshape(-1, 1) # odómetro como feature
y_reg = df_clean['future_price_5yr'].values      # precio futuro (target de Regresión)
```

**Modelo de Regresión - Predicción de Precio Futuro**

Este modelo predice el valor del precio en 5 años, basándose únicamente en el odómetro actual.

Este modelo utiliza la Regresión Lineal para predecir el valor exacto del automóvil en 5 años (simulado con el 50% del precio actual) basándose en la lectura del odómetro.

Error Absoluto Medio - MAE: En promedio, el precio predicho por el modelo se desvía del precio real (simulado) en $2,739.78.

Coeficiente de Determinación R^2 : El 33.87% de la variación del precio futuro simulado puede ser explicada por la lectura del odómetro. (Este valor indica una correlación moderada).

División de datos y entrenamiento

```python
# División para Regresión
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.3, random_state=42
)

# Inicializar y entrenar el modelo de Regresión Lineal
model_reg = LinearRegression()
model_reg.fit(X_train_reg, y_train_reg)

print("Modelo de Regresión Lineal entrenado.")
print("\nDimensiones de los conjuntos de regresión:")
print("X_train_reg:", X_train_reg.shape, "y_train_reg:", y_train_reg.shape)
print("X_test_reg:", X_test_reg.shape, "y_test_reg:", y_test_reg.shape)
```

Predicción y Evaluación

```python
# Predicciones
y_pred_reg = model_reg.predict(X_test_reg)

# Evaluación con métricas de Regresión
mae = mean_absolute_error(y_test_reg, y_pred_reg)
r2 = r2_score(y_test_reg, y_pred_reg)

print("\n--- Resultados de Regresión ---")
print(f"Error Absoluto Medio (MAE): ${mae:,.2f}")
print(f"Coeficiente de Determinación (R²): {r2:.4f}")


# Comparar algunas predicciones con los valores reales
print("\nAlgunas predicciones vs valores reales:")
comparison_df = pd.DataFrame({'Real': y_test_reg[:10], 'Predicho': y_pred_reg[:10]})
print(comparison_df)

# Visualización de la Regresión
plt.figure(figsize=(8, 6))
plt.scatter(X_test_reg, y_test_reg, color='blue', label='Valores Reales (Simulado)')
plt.plot(X_test_reg, y_pred_reg, color='red', linewidth=2, label='Línea de Predicción')
plt.title('Regresión Lineal: Precio Estimado a 5 Años vs. Odométer')
plt.xlabel('Odométer (millas)')
plt.ylabel('Precio Estimado a 5 Años ($)')
plt.legend()
plt.show()
```

**Modelo de Clasificación**

Para usar las métricas (accuracy, matriz de confusión), se crea la variable objetivo de clasificación binaria.

Ingenieria de Caracteristicas

Se clasifican los autos en 'Alto Precio' o 'Bajo Precio' basándonos en si su future_price_5yr está por encima o por debajo de la mediana.

```python
# Calcular la mediana del precio futuro simulado
median_price = df_clean['future_price_5yr'].median()

# Crear la variable binaria: 1 si es 'Caro' (por encima de la mediana), 0 si es 'Barato'
df_clean['price_class'] = (df_clean['future_price_5yr'] > median_price).astype(int)

# Definir el nuevo target de Clasificación
y_class = df_clean['price_class'].values

print(f"Mediana del Precio Futuro: ${median_price:,.2f}")
print(f"Distribución de la Clase (0=Barato, 1=Caro):\n{df_clean['price_class'].value_counts()}")
```

División de Datos y Entrenamiento

```python
# División para Clasificación
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X, y_class, test_size=0.3, random_state=42, stratify=y_class # stratify para balancear clases
)

# Inicializar y entrenar el modelo de Regresión Logística (comúnmente usado para clasificación)
model_cls = LogisticRegression(max_iter=200)
model_cls.fit(X_train_cls, y_train_cls)

print("Modelo de Clasificación (Regresión Logística) entrenado.")
```

Predicciones y Evaluación (Métricas de Clasificación)

Este modelo utiliza la Regresión Logística para clasificar el precio futuro del automóvil como 0 (Barato) o 1 (Caro), basándose en si está por debajo o por encima de la mediana.

El modelo clasificó correctamente el 77.78% de los automóviles como 'Baratos' o 'Caros'.

La matriz de confusión detalla el rendimiento del modelo, comparando las etiquetas reales con las predichas.
60122 (Verdaderos Negativos)
70245 (Verdaderos Positivos)

Precisión (Precision): De todos los autos que el modelo predijo como 'Caros' (Clase 1), el 75% realmente lo eran.

Recall (Sensibilidad): De todos los autos que realmente eran 'Caros' (Clase 1), el modelo identificó correctamente el 82%.

**Conclusión**: El modelo de clasificación tuvo un buen rendimiento general Accuracy=0.7778. Es mejor identificando correctamente los autos 'Caros' (Recall del $0.84$) que prediciendo correctamente los autos 'Baratos' (Precisión del $0.84$). Esto es útil para un comprador, ya que el modelo tiene una alta capacidad para identificar los vehículos que mantendrán un valor superior a la mediana.

```python
# Predicciones
y_pred_cls = model_cls.predict(X_test_cls)

print("\n--- Resultados de Clasificación ---")

# 1. Accuracy
accuracy = accuracy_score(y_test_cls, y_pred_cls)
print(f"Accuracy (Precisión General): {accuracy:.4f}")

# 2. Matriz de Confusión
cm = confusion_matrix(y_test_cls, y_pred_cls)
print("\nMatriz de Confusión:")
print(cm)
# Visualización de la Matriz de Confusión
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Barato (0)', 'Caro (1)'],
            yticklabels=['Barato (0)', 'Caro (1)'])
plt.ylabel('Etiquetas Reales')
plt.xlabel('Etiquetas Predichas')
plt.title('Matriz de Confusión')
plt.show()
#

# 3. Reporte de Clasificación (Precision, Recall, F1-Score)
report = classification_report(y_test_cls, y_pred_cls, target_names=['Barato (0)', 'Caro (1)'])
print("\nReporte de Clasificación:")
print(report)
```
